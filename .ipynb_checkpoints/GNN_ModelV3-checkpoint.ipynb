```python
import torch
import torch.nn.functional as F
from torch_geometric.loader import DataLoader
from torch_geometric.nn import GCNConv, global_mean_pool, BatchNorm
from torch_geometric.data import Data
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.utils.class_weight import compute_class_weight
from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import RandomUnderSampler
from sklearn.metrics import confusion_matrix, f1_score
import matplotlib.pyplot as plt
import seaborn as sns

# 🚀 Load datasets
print("→ Loading datasets...")
df1_full = pd.read_csv("Z2JetsToNuNu/Z2JetsToNuNu1_PtZ-150To250.csv")
df2_full = pd.read_csv("Z2JetsToNuNu_Negative/ZToNuNu_PtZ-150To250.csv")

df1_full["target"] = 1  
df2_full["target"] = 0  

df_full = pd.concat([df1_full, df2_full], ignore_index=True)
print(f"→ Data loaded! Total events: {df_full.shape[0]}")

df_full = df_full.apply(pd.to_numeric, errors="coerce")
df_full.fillna(0, inplace=True)

# 🔹 Apply balancing techniques
undersample = RandomUnderSampler(sampling_strategy=0.4, random_state=42)
X_under, y_under = undersample.fit_resample(df_full.drop(columns=["target"]).values, df_full["target"].values)

smote = SMOTE(sampling_strategy=0.6, random_state=42)
X_bal, y_bal = smote.fit_resample(X_under, y_under)

df_bal = pd.DataFrame(X_bal, columns=df_full.columns[:-1])
df_bal["target"] = y_bal

print(f"→ After balancing: {df_bal['target'].value_counts().to_dict()}")

jet_columns = ["lead_jet_pt", "lead_jet_eta", "lead_jet_phi", "lead_jet_mass"]

# 🔹 Function to build the graph representation
def build_graph(event):
    nodes = [float(event[col]) for col in jet_columns]
    edges = [[i, i + 1] for i in range(len(nodes) - 1)]
    
    x = torch.tensor(nodes, dtype=torch.float).view(-1, 1)
    edge_index = torch.tensor(edges, dtype=torch.long).t().contiguous()
    
    return Data(x=x, edge_index=edge_index, y=torch.tensor([event["target"]], dtype=torch.long))

graph_data = [build_graph(row) for _, row in df_bal.iterrows()]
train_data, test_data = train_test_split(graph_data, test_size=0.2, random_state=42)

train_loader = DataLoader(train_data, batch_size=64, shuffle=True)
test_loader = DataLoader(test_data, batch_size=64, shuffle=False)

y_train = [data.y.item() for data in train_data]
class_weights = compute_class_weight(class_weight="balanced", classes=np.unique(y_train), y=y_train)
class_weights = torch.tensor(class_weights, dtype=torch.float)
class_weights[1] *= 1.5  # 🔥 Adjusted to reduce false positives

# 🔹 Define GNN Model
class GNN(torch.nn.Module):
    def __init__(self):
        super(GNN, self).__init__()
        self.conv1 = GCNConv(1, 16)
        self.bn1 = BatchNorm(16)
        self.conv2 = GCNConv(16, 32)
        self.bn2 = BatchNorm(32)
        self.conv3 = GCNConv(32, 64)
        self.bn3 = BatchNorm(64)
        self.conv4 = GCNConv(64, 16)
        self.bn4 = BatchNorm(16)
        self.dropout = torch.nn.Dropout(p=0.4)  # 🔥 Reduced dropout for stability
        self.fc = torch.nn.Linear(16, 2)

    def forward(self, data):
        x, edge_index = data.x, data.edge_index
        x = self.dropout(F.relu(self.bn1(self.conv1(x, edge_index))))
        x = self.dropout(F.relu(self.bn2(self.conv2(x, edge_index))))
        x = self.dropout(F.relu(self.bn3(self.conv3(x, edge_index))))
        x = self.dropout(F.relu(self.bn4(self.conv4(x, edge_index))))
        x = global_mean_pool(x, data.batch) if hasattr(data, "batch") else x.mean(dim=0, keepdim=True)
        return F.softmax(self.fc(x), dim=1)  # 🔥 Now outputs probabilities

# 🔹 Define training parameters
model = GNN()
optimizer = torch.optim.Adam(model.parameters(), lr=0.0005, weight_decay=1e-3)
scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode='max', patience=2, factor=0.5)
criterion = torch.nn.CrossEntropyLoss(weight=class_weights)

# 🚀 Train for 10 epochs
epochs = 10  
accuracy_list, f1_list, loss_list = [], [], []

for epoch in range(epochs):
    model.train()
    total_loss = 0

    for data in train_loader:
        optimizer.zero_grad()
        out = model(data)
        loss = criterion(out, data.y)
        loss.backward()
        torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)
        optimizer.step()
        total_loss += loss.item()

    # 🔹 Evaluate using a dynamic threshold
    model.eval()
    y_true, y_pred = [], []
    with torch.no_grad():
        for data in test_loader:
            out = model(data)
            probs = out[:, 1]  # 🔥 Taking only class 1 probability
            pred = (probs > 0.6).long()  # 🔥 Dynamic threshold set at 0.6
            y_true.extend(data.y.cpu().numpy())
            y_pred.extend(pred.cpu().numpy())

    acc = np.mean(np.array(y_true) == np.array(y_pred)) * 100
    f1 = f1_score(y_true, y_pred, average="macro")
    avg_loss = total_loss / len(train_loader)

    accuracy_list.append(acc)
    f1_list.append(f1)
    loss_list.append(avg_loss)

    print(f"→ Epoch {epoch+1}/{epochs}, Accuracy: {acc:.2f}%, F1-Score: {f1:.4f}, Loss: {avg_loss:.4f}")

print("→ Training completed at epoch 10!")

# 🚀 Plot and save results
plt.figure(figsize=(10, 5))
plt.plot(range(1, len(accuracy_list) + 1), accuracy_list, marker='o', linestyle='-', label="Accuracy", color='blue')
plt.plot(range(1, len(f1_list) + 1), np.array(f1_list) * 100, marker='s', linestyle='--', label="F1-Score (%)", color='green')
plt.xlabel("Epoch")
plt.ylabel("Accuracy / F1-Score (%)")
plt.title("Accuracy and F1-Score Over Training")
plt.legend()
plt.grid(True)
plt.savefig("accuracy_f1score_v3.png", dpi=300)
plt.show()

plt.figure(figsize=(10, 5))
plt.plot(range(1, len(loss_list) + 1), loss_list, marker='o', linestyle='-', color='purple', label='Loss')
plt.xlabel("Epoch")
plt.ylabel("Loss (CrossEntropyLoss)")
plt.title("Loss Over Training")
plt.legend()
plt.grid(True)
plt.savefig("loss_training_v3.png", dpi=300)
plt.show()

cm = confusion_matrix(y_true, y_pred)
cm_df = pd.DataFrame(cm, index=["Class 0", "Class 1"], columns=["Predicted 0", "Predicted 1"])
plt.figure(figsize=(6, 5))
sns.heatmap(cm_df, annot=True, fmt="d", cmap="Blues", cbar=False)
plt.title("Final Confusion Matrix")
plt.ylabel("True Label")
plt.xlabel("Predicted Label")
plt.savefig("confusion_matrix_v3.png", dpi=300)
plt.show()

test_indices = df_bal.index[len(train_data):]  
df_predictions = df_bal.iloc[test_indices].copy()
df_predictions["predicted_class"] = y_pred 

df_filtered = df_predictions[
    (df_predictions["predicted_class"] == 1) & (df_predictions["recoil_magnitude_u"] > 200)
]

df_filtered.to_csv("filtered_events.csv", index=False)
```
