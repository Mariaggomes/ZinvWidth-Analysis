```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import ast
```

```python
df_ML = pd.read_csv("../test_data/Z2JetsToNuNu_class1.csv")
df_ML.head()
```

```python
df_cuts = pd.read_csv("../test_data/selected_events_Z2JetsToNuNu_PtZ-250To400.csv")
df_cuts.head()
```

```python
print(len(df_ML))
print(len(df_cuts))
```

```python
#Recoil Magnitude U Distribution
plt.figure(figsize=(10, 6))

# ML
plt.hist(df_ML["recoil_magnitude_u"], bins=50, range=(0, 800), color='blue', alpha=0.4,
         edgecolor='black', linewidth=0.5, histtype='stepfilled', label='ML')
# Cuts
plt.hist(df_cuts["recoil_magnitude_u"], bins=50, range=(0, 800), color='red', alpha=0.4,
         edgecolor='black', linewidth=0.5, histtype='stepfilled', label='Cuts')

plt.title('U Distribution (0–800 GeV)', fontsize=20)
plt.xlabel('U [GeV]', fontsize=16)
plt.ylabel('Frequency', fontsize=16)
plt.legend(fontsize=14)
plt.grid(False)
plt.tight_layout()
plt.savefig("../plots/recoil_magnitude_u_distribution.png", dpi=300)
plt.show()
```

```python
# MET Distribution
plt.figure(figsize=(10, 6))

#ML
plt.hist(df_ML["met_pt"], bins=50, color='blue', alpha=0.4,
         edgecolor='black', linewidth=0.5, histtype='stepfilled', label='ML')

#Cuts
plt.hist(df_cuts["met_pt"], bins=50, color='red', alpha=0.4,
         edgecolor='black', linewidth=0.5, histtype='stepfilled', label='Cuts')

plt.title('MET Distribution', fontsize=20)
plt.xlabel('MET [GeV]', fontsize=16)
plt.ylabel('Frequency', fontsize=16)
plt.legend(fontsize=14)
plt.grid(False)
plt.tight_layout()
plt.savefig("../plots/met_distribution.png", dpi=300)
plt.show()
```

```python
# Lead Jet pt Distribution
plt.figure(figsize=(10, 6))

# ML
plt.hist(df_ML["lead_jet_pt"], bins=50, color='blue', alpha=0.4,
         edgecolor='black', linewidth=0.5, histtype='stepfilled', label='ML')

# Cuts
plt.hist(df_cuts["lead_jet_pt"], bins=50, color='red', alpha=0.4,
         edgecolor='black', linewidth=0.5, histtype='stepfilled', label='Cuts')

plt.title('Lead Jet pt Distribution', fontsize=20)
plt.xlabel('Lead Jet pt [GeV]', fontsize=16)
plt.ylabel('Frequency', fontsize=16)
plt.legend(fontsize=14)
plt.grid(False)  
plt.tight_layout()
plt.savefig("../plots/lead_jet_pt_distribution.png", dpi=300)
plt.show()
```

```python
# Lead Jet eta Distribution
plt.figure(figsize=(10, 6))

# ML
plt.hist(df_ML["lead_jet_eta"], bins=50, color='blue', alpha=0.4,
         edgecolor='black', linewidth=0.5, histtype='stepfilled', label='ML')

# Cuts
plt.hist(df_cuts["lead_jet_eta"], bins=50, color='red', alpha=0.4,
         edgecolor='black', linewidth=0.5, histtype='stepfilled', label='Cuts')

xticks = list(plt.xticks()[0]) + [-2.4, 2.4]
xticks = sorted(set(xticks))  
plt.xticks(xticks)

plt.title('Lead Jet η Distribution', fontsize=20)
plt.xlabel('Lead Jet η', fontsize=16)
plt.ylabel('Frequency', fontsize=16)
plt.legend(fontsize=14)
plt.grid(False)
plt.tight_layout()
plt.savefig("../plots/lead_jet_eta_distribution.png", dpi=300)
plt.show()
```

```python
# Lead Jet ch hef Distribution
plt.figure(figsize=(10, 6))

# ML
plt.hist(df_ML["lead_jet_ch_hef"], bins=50, color='blue', alpha=0.4,
         edgecolor='black', linewidth=0.5, histtype='stepfilled', label='ML')

# Cuts
plt.hist(df_cuts["lead_jet_ch_hef"], bins=50, color='red', alpha=0.4,
         edgecolor='black', linewidth=0.5, histtype='stepfilled', label='Cuts')

xticks = list(plt.xticks()[0]) + [0.95]
xticks = sorted(set(xticks))
plt.xticks(xticks)

plt.title('Lead Jet Charged Hadron Energy Fraction (chHEF) Distribution', fontsize=20)
plt.xlabel('Lead Jet chHEF', fontsize=16)
plt.ylabel('Frequency', fontsize=16)
plt.legend(fontsize=14)
plt.grid(False)
plt.tight_layout()
plt.savefig("../plots/lead_jet_chhef_distribution.png", dpi=300)
plt.show()
```

```python
# Convert to lists
df_ML["photon_pt"] = df_ML["photon_pt"].apply(ast.literal_eval)
df_cuts["photon_pt"] = df_cuts["photon_pt"].apply(ast.literal_eval)

# Flatten the lists 
photon_pt_ML = [pt for row in df_ML["photon_pt"] for pt in row]
photon_pt_cuts = [pt for row in df_cuts["photon_pt"] for pt in row]
```

```python
# Photon pt Distribution
bins = np.linspace(0, 500, 50)
plt.figure(figsize=(10, 6))

#ML
plt.hist(photon_pt_ML, bins=bins, color='blue', alpha=0.4,
         edgecolor='black', linewidth=0.5, histtype='stepfilled', label='ML')

#Cuts
plt.hist(photon_pt_cuts, bins=bins, color='red', alpha=0.4,
         edgecolor='black', linewidth=0.5, histtype='stepfilled', label='Cuts')

plt.title('Photon pt Distribution', fontsize=20)
plt.xlabel('Photon_pt [GeV]', fontsize=16)
plt.ylabel('Frequency', fontsize=16)
plt.legend(fontsize=14)
plt.grid(False)
plt.tight_layout()
plt.savefig("../plots/photon_pt_distribution.png", dpi=300)
plt.show()
```
