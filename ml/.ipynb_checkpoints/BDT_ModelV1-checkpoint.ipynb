```python
import pandas as pd
import numpy as np
import xgboost as xgb
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, roc_curve, auc
from imblearn.over_sampling import SMOTE
from collections import Counter

# 🔹 Carregar os datasets
file1_path = "Z2JetsToNuNu/Z2JetsToNuNu1_PtZ-150To250.csv"
file2_path = "Z2JetsToNuNu_Negative/ZToNuNu_PtZ-150To250.csv"

df1_full = pd.read_csv(file1_path)
df2_full = pd.read_csv(file2_path)

# 🔹 Adicionar rótulos
df1_full["target"] = 1  # Eventos de interesse
df2_full["target"] = 0  # Eventos de fundo

# 🔹 Concatenar os datasets
df_full = pd.concat([df1_full, df2_full], ignore_index=True)

# 🔹 Remover colunas não numéricas e preencher valores NaN
df_full = df_full.apply(pd.to_numeric, errors='coerce')
df_full = df_full.fillna(0)  

# 🔹 Separar features e target
X = df_full.drop(columns=["target"]).values
y = df_full["target"].values

# 🔹 Aplicar apenas SMOTE para balanceamento
smote = SMOTE(sampling_strategy=0.6, random_state=42)
X_bal, y_bal = smote.fit_resample(X, y)

# 🔹 Divisão treino/teste
X_train, X_test, y_train, y_test = train_test_split(X_bal, y_bal, test_size=0.2, stratify=y_bal, random_state=42)

# 🔹 Ajustar `scale_pos_weight`
num_class_0, num_class_1 = Counter(y_train).values()
scale_pos_weight = num_class_0 / num_class_1

# 🔹 Criar modelo XGBoost com ajustes finais e DART
model_bdt = xgb.XGBClassifier(
    booster="dart",
    n_estimators=40,
    max_depth=2,
    learning_rate=0.03,
    gamma=50.0,
    reg_lambda=10.0,
    reg_alpha=8.0,
    min_child_weight=20,
    subsample=0.1,
    colsample_bytree=0.1,
    max_delta_step=2,
    scale_pos_weight=scale_pos_weight,
    use_label_encoder=False,
    eval_metric="logloss"
)

# 🔹 Treinar o modelo com early stopping
eval_set = [(X_train, y_train), (X_test, y_test)]
model_bdt.fit(X_train, y_train, early_stopping_rounds=10, eval_set=eval_set, verbose=True)

# 🔹 Fazer previsões no conjunto de teste
y_pred = model_bdt.predict(X_test)
y_prob = model_bdt.predict_proba(X_test)[:, 1]

# 🔹 Avaliação do modelo
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)

# 🔹 Criar gráficos de validação
fig, ax = plt.subplots(1, 3, figsize=(18, 5))

# 🔹 Gráfico 1: Evolução do LogLoss
logloss_values = model_bdt.evals_result()["validation_0"]["logloss"]
ax[0].plot(range(len(logloss_values)), logloss_values, label="Validation LogLoss")
ax[0].set_xlabel("Iterações")
ax[0].set_ylabel("LogLoss")
ax[0].set_title("Evolução do LogLoss")
ax[0].legend()

# 🔹 Gráfico 2: Matriz de Confusão
sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Blues",
            xticklabels=["Classe 0", "Classe 1"], yticklabels=["Classe 0", "Classe 1"], ax=ax[1])
ax[1].set_xlabel("Predito")
ax[1].set_ylabel("Real")
ax[1].set_title("Matriz de Confusão")

# 🔹 Gráfico 3: Curva ROC
fpr, tpr, _ = roc_curve(y_test, y_prob)
roc_auc = auc(fpr, tpr)
ax[2].plot(fpr, tpr, label=f"ROC Curve (AUC = {roc_auc:.2f})")
ax[2].plot([0, 1], [0, 1], linestyle="--")
ax[2].set_xlabel("Falso Positivo Rate")
ax[2].set_ylabel("Verdadeiro Positivo Rate")
ax[2].set_title("Curva ROC")
ax[2].legend()

plt.tight_layout()
plt.show()

# 🔹 Exibir resultados
print(f"\n **Resultados do Modelo BDT Ajustado (com DART)**")
print(f"Acurácia: {accuracy * 100:.2f}%")
print(report)
```
