```python
all_mc_pileup_values = []
all_mc_weights = []
```

```python
for dataset in datasets:
    # Initializes variables to store cumulative values.
    N_gen = 0
    gw_pos = 0
    gw_neg = 0
    # Flag to indicate whether the current dataset is collision.
    IS_COLLISION = False

    path = r'./'
    # Get a list of all CSV files that match the current dataset name
    all_files = glob.glob(os.path.join(path, f"{dataset}.csv"))
    print(f"Processing {len(all_files)} files in {dataset}")

    list_of_dataframes = []
    # Loop over each file in the dataset
    for filename in all_files:
        # Checks if the file is a collision dataset.
        if 'collision' in filename:
            IS_COLLISION = True
        # Read the CSV file into a DataFrame
        df = pd.read_csv(filename, index_col=None, header=0)
        list_of_dataframes.append(df)

        # For simulated data (non-collision), accumulate total values for normalization
        if not IS_COLLISION and len(df) > 0:
            N_gen += df['N_gen'][0]
            gw_pos += df['gw_pos'][0]
            gw_neg += df['gw_neg'][0]

    if not IS_COLLISION and list_of_dataframes:
        # Concatenate all DataFrames into a single one
        df_all = pd.concat(list_of_dataframes, ignore_index=True)
        # Extract pileup values and compute histogram
        pileup_vals = df_all['pileup'].values
        mc_values, _ = np.histogram(pileup_vals, bins=bins)
        # Normalize the Monte Carlo histogram
        mc_values = mc_values / np.sum(mc_values)
        
        # Compute pileup reweighting factors: data / MC
        weights_pileup = np.divide(data_values, mc_values, out=np.ones_like(data_values), where=mc_values != 0)
        # Get indices of the bins for each event's pileup value
        indices = np.digitize(pileup_vals, bins) - 1
        indices = np.clip(indices, 0, len(weights_pileup) - 1)
        # Assign weights to each event based on its pileup
        df_all['pileup_weight'] = weights_pileup[indices]

        # Store pileup and weight values for further use or statistics
        all_mc_pileup_values.extend(pileup_vals)
        all_mc_weights.extend(weights_pileup[indices])
        
        # Add total event count and weights to the DataFrame
        df_all['N_gen'] = N_gen
        df_all['gw_pos'] = gw_pos
        df_all['gw_neg'] = gw_neg

    elif IS_COLLISION and list_of_dataframes:
        # Concatenate the DataFrames
        df_all = pd.concat(list_of_dataframes, ignore_index=True)
        # Fill placeholders for non-applicable fields in collision data
        df_all['N_gen'] = -999
        df_all['gw_pos'] = -999
        df_all['gw_neg'] = -999
    # If no files were found for the dataset
    else:
        # Create an empty DataFrame with the expected structure
        df_all = pd.DataFrame(columns=['N_gen', 'nevents', 'gw_pos', 'gw_neg'])
        print(f"No files found for {dataset}. Empty DataFrame created.")
    
    output_path = '../data/processed_datasets'
    # Ensure the output directory exists
    os.makedirs(output_path, exist_ok=True)

    # Extract dataset name and save the processed DataFrame to a new CSV file
    dataset_name = os.path.basename(dataset)
    df_all.to_csv(os.path.join(output_path, f'Processed_{dataset_name}.csv'), index=False)
```

```python
# Compute bin centers 
bin_centers = 0.5 * (bins[:-1] + bins[1:])

# Real data pileup histogram 
plt.hist(
    bins[:-1], bins=bins, weights=data_values,
    histtype='stepfilled', alpha=0.5, label='Data (PU)', color='black'
)

# Raw MC as curve
mc_hist_raw, _ = np.histogram(all_mc_pileup_values, bins=bins)
mc_hist_raw = mc_hist_raw / np.sum(mc_hist_raw)
plt.plot(bin_centers, mc_hist_raw, label='MC (raw)', color='red', linestyle='--', linewidth=2)

# KDE-smoothed MC weighted distribution
sns.kdeplot(
    x=all_mc_pileup_values,
    weights=all_mc_weights,
    bw_adjust=1.2,
    label='MC (weighted)', color='blue', linewidth=2
)

plt.xlabel('Pileup (n_interactions)')
plt.ylabel('Normalized Events')
plt.title('Pileup Distribution: Data vs MC')
plt.legend()
plt.tight_layout()
plt.savefig("../plots/pileup_comparison.png", dpi=300)
plt.show()
```

```python
# Correction factor for gamma* (only for DYJets datasets)
gamma_star_correction = 0.9519 

background_datasets = [
    # Z → νν backgrounds 
    r"Z2JetsToNuNu*_PtZ-50To150", r"Z2JetsToNuNu_PtZ-150To250_*", r"Z2JetsToNuNu_PtZ-250To400_*", 
    r"Z2JetsToNuNu_PtZ-400ToInf_*", r"Z1JetsToNuNu*_PtZ-150To250", r"Z1JetsToNuNu*_PtZ-250To400",
    r"Z1JetsToNuNu*_PtZ-400ToInf",
    
    # --- W → lν backgrounds 
    r"WJetsToLNu*_Pt-150To250-Ptmiss", r"WJetsToLNu*_Pt-250To400-Ptmiss",
    r"WJetsToLNu*_Pt-400To600-Ptmiss", r"WJetsToLNu*_Pt-600ToInf-Ptmiss",
    r"WJetsToLNu*_Pt-100To250-SingleElectron", r"WJetsToLNu*_Pt-250To400-SingleElectron",
    r"WJetsToLNu*_Pt-400To600-SingleElectron", r"WJetsToLNu*_Pt-600ToInf-SingleElectron",
    r"WJetsToLNu*_Pt-100To250-SingleMuon", r"WJetsToLNu*_Pt-250To400-SingleMuon",
    r"WJetsToLNu*_Pt-400To600-SingleMuon", r"WJetsToLNu*_Pt-600ToInf-SingleMuon",

    # --- WZ backgrounds 
    r"WZTo1L1Nu2Q_SingleElectron*", r"WZTo1L1Nu2Q_SingleMuon*", r"WZTo1L1Nu2Q_SingleTau*",
    r"WZTo1L3Nu_SingleElectron*", r"WZTo1L3Nu_SingleMuon*", r"WZTo1L3Nu_SingleTau*",
    r"WZTo3LNu_SingleElectron*", r"WZTo3LNu_SingleMuon*", r"WZTo3LNu_SingleTau*",
    r"WZTo1L3Nu_Ptmiss*",

    # WW backgrounds
    r"WWTo1L1Nu2Q_SingleElectron*", r"WWTo1L1Nu2Q_SingleMuon*", r"WWTo1L1Nu2Q_SingleTau*",
    r"WWTo2L2Nu_SingleElectron*", r"WWTo2L2Nu_SingleMuon*", r"WWTo2L2Nu_SingleTau*",
    r"WWTo2L2Nu_Ptmiss*",

    # ZZ backgrounds 
    r"ZZTo2L2Nu_SingleElectron*", r"ZZTo2L2Nu_SingleMuon*", r"ZZTo2L2Nu_SingleTau*",
    r"ZZTo2L2Nu_Ptmiss*",

    # Top pair production 
    r"TTJets_SingleElectron*", r"TTJets_SingleMuon*", r"TTJets_SingleTau*",

    # W + γ backgrounds 
    r"WGToLNuG_SingleElectron*", r"WGToLNuG_SingleMuon*", r"WGToLNuG_SingleTau*",
    r"ZGTo2NuG_Ptmiss*",

    # Electroweak W production
    r"EWKWMinus_SingleElectron*", r"EWKWMinus_SingleMuon*", r"EWKWMinus_SingleTau*",
    r"EWKWPlus_SingleElectron*", r"EWKWPlus_SingleMuon*", r"EWKWPlus_SingleTau*",
    r"EWKZ2Jets_Ptmiss*",

    # Single Top production 
    r"ST_tW_antitop_SingleElectron*", r"ST_tW_antitop_SingleMuon*", r"ST_tW_antitop_SingleTau*",
    r"ST_tW_top_SingleElectron*", r"ST_tW_top_SingleMuon*", r"ST_tW_top_SingleTau*",
    r"ST_t-channel_antitop_SingleElectron*", r"ST_t-channel_antitop_SingleMuon*", r"ST_t-channel_antitop_SingleTau*",
    r"ST_t-channel_top_SingleElectron*", r"ST_t-channel_top_SingleMuon*", r"ST_t-channel_top_SingleTau*",

    # Drell-Yan 
    r"DYJetsToLL*_PtZ-250To400", r"DYJetsToLL*_PtZ-400To650", r"DYJetsToLL*_PtZ-650ToInf"
]

# Calculate weights for all background samples
for dataset in background_datasets:
    N_gen = data[dataset]['N_gen']
    xsec = data[dataset]['xsec']

    # Base weight: luminosity * cross section / number of generated events
    base_weight = integrated_luminosity * xsec / N_gen

    # Apply gamma* correction only for DYJetsToLL datasets
    if "DYJetsToLL" in dataset:
        weight = base_weight * gamma_star_correction
    else:
        weight = base_weight
        
    # Save the weight
    data[dataset]['weight'] = weight
    
    # Apply pileup reweighting
    df['weight'] = weight * df['pileup_weight']


# Collision (data) 
# All data samples are assigned a weight of 1
collision_datasets = [
    'Met*_collision', 'MetSingleMuon*_collision', 'MetSingleTau*_collision',
    'SingleElectron*_collision', 'SingleMuon*_collision',
    'DoubleMuon-DoubleMuon*_collision', 'Met-DoubleMuon*_collision',
    'SingleMuon-DoubleMuon*_collision', 'DoubleMuon-DoubleMuon*_collision',
]

for dataset in collision_datasets:
    data[dataset]['weight'] = 1
```

```python
# transfer factors
def ratio_with_error(numerator, denominator):
    if denominator == 0 or numerator == 0:
        return 0.0, 0.0
    ratio = numerator / denominator
    error = ratio * np.sqrt(1/numerator + 1/denominator)
    return ratio, error

# Transfer factor r_Z (Z → νν from Z ll)
r_Z, err_r_Z = ratio_with_error(s_Zj.sum(), s_Zll.sum())

# Transfer factor r_W (W → lν from region 1 lepton)
r_W, err_r_W = ratio_with_error(b_Wj.sum(), b_Wl.sum())

print(f"rZ = {r_Z:.4f} ± {err_r_Z:.4f}, rW = {r_W:.4f} ± {err_r_W:.4f}")
           
# Systematic uncertainties (converted to fraction)
syst_uncertainty = 0.0329  # 3.29% of the CMS Table
sigma_theta = syst_uncertainty
```

MCMC (Markov Chain Monte Carlo) is a sampling technique used to explore complex probability distributions, particularly in models with multiple correlated parameters. In this analysis, MCMC is employed to estimate the posterior distributions of the model parameters (r, rZ, and θ), given the observed data and prior uncertainties from theory and experimental inputs. By generating a chain of samples that follow the joint probability distribution, MCMC allows us to extract not only the best-fit values of the parameters but also their credible intervals and correlations. This enables a robust determination of the invisible width of the Z boson, incorporating all relevant sources of uncertainty in a statistically consistent way.

```python
# Penalties
sigma_theta = 0.032  # global systematic uncertainty
sigma_r = 0.1         # informative prior for r
sigma_rZ = 0.2        # informative prior for rZ

def log_prob(params):
    theta, r, rZ = params
    if not (-1 < theta < 1 and 0 < r < 5 and 0.01 < rZ < 10):
        return -np.inf

    epsilon = 1e-10
    try:
        mu_j = r * rZ * s_Zj * (1 + theta) + r_W * b_Wj * (1 + theta) + b_bkg_j
        mu_l = r_W * b_Wl * (1 + theta) + b_bkg_l
        mu_ll = rZ * s_Zll * (1 + theta) + np.sqrt(rZ) * (s_int + s_gamma)

        mu_j = np.clip(mu_j, epsilon, None)
        mu_l = np.clip(mu_l, epsilon, None)
        mu_ll = np.clip(mu_ll, epsilon, None)

        penalty_theta = norm.logpdf(theta, 0, sigma_theta)
        penalty_r = norm.logpdf(r, 1.0, sigma_r)
        penalty_rZ = norm.logpdf(rZ, 0.7, sigma_rZ)

        logL = (
            np.sum(poisson.logpmf(n_j, mu_j)) +
            np.sum(poisson.logpmf(n_l, mu_l)) +
            np.sum(poisson.logpmf(n_ll, mu_ll)) +
            penalty_theta + penalty_r + penalty_rZ
        )

        return logL if np.isfinite(logL) else -np.inf

    except Exception:
        return -np.inf

# RUN MCMC
ndim = 3
nwalkers = 40
nsteps = 5000
initial = np.array([0.0, 1.0, 1.0])  # theta, r, rZ
pos = initial + 1e-2 * np.random.randn(nwalkers, ndim)

sampler = emcee.EnsembleSampler(nwalkers, ndim, log_prob)
sampler.run_mcmc(pos, nsteps, progress=True)

# EXTRACT POSTERIOR SAMPLES
burnin = 1000
samples = sampler.get_chain(discard=burnin, flat=True)

theta_samples = samples[:, 0]
r_samples = samples[:, 1]
rZ_samples = samples[:, 2]

# ---------- STATISTICS FOR r ----------
r_median = np.median(r_samples)
r_low = np.percentile(r_samples, 16)
r_high = np.percentile(r_samples, 84)
r_err_down = r_median - r_low
r_err_up = r_high - r_median

# ---------- STATISTICS FOR rZ ----------
rZ_median = np.median(rZ_samples)
rZ_low = np.percentile(rZ_samples, 16)
rZ_high = np.percentile(rZ_samples, 84)
rZ_err_down = rZ_median - rZ_low
rZ_err_up = rZ_high - rZ_median

# ---------- STATISTICS FOR theta ----------
theta_median = np.median(theta_samples)
theta_low = np.percentile(theta_samples, 16)
theta_high = np.percentile(theta_samples, 84)
theta_err_down = theta_median - theta_low
theta_err_up = theta_high - theta_median

# COMPUTE INVISIBLE Z WIDTH
Gamma_SM = 499.0  # MeV
Gamma_samples = r_samples * Gamma_SM

Gamma_median = np.median(Gamma_samples)
Gamma_low = np.percentile(Gamma_samples, 16)
Gamma_high = np.percentile(Gamma_samples, 84)
Gamma_err_down = Gamma_median - Gamma_low
Gamma_err_up = Gamma_high - Gamma_median

# ---------- PRINT ALL RESULTS ----------
print(f"Γ(Z → νν̄) = {Gamma_median:.1f} +{Gamma_err_up:.1f} / -{Gamma_err_down:.1f} MeV")
print(f"r         = {r_median:.3f} +{r_err_up:.3f} / -{r_err_down:.3f}")
print(f"rZ        = {rZ_median:.3f} +{rZ_err_up:.3f} / -{rZ_err_down:.3f}")
print(f"theta     = {theta_median:.4f} +{theta_err_up:.4f} / -{theta_err_down:.4f}")

# GENERATE CORNER PLOT AND SAVE
labels = [r"$\theta$", r"$r$", r"$r_Z$", r"$\Gamma_{inv}$"]
all_samples = np.column_stack([theta_samples, r_samples, rZ_samples, Gamma_samples])
fig = corner.corner(all_samples, labels=labels, truths=[None, None, None, Gamma_median])
plot_path = "../plots/corner_plot_GammaInv.png"
fig.savefig(plot_path)
```

```python
# Known uncertainty on rW (fixed parameter)
sigma_rW = 0.0180  # absolute uncertainty of rW
rW_central = 1.4050  # fixed central value used in the fit

# Number of MCMC samples
N = len(r_samples)

# Recalculate Gamma_inv (invisible Z width) from MCMC samples
Gamma_samples = r_samples * Gamma_SM  # already done in MCMC

# Compute median and quantiles from MCMC
Gamma_median = np.median(Gamma_samples)
Gamma_low = np.percentile(Gamma_samples, 16)
Gamma_high = np.percentile(Gamma_samples, 84)
Gamma_err_down = Gamma_median - Gamma_low
Gamma_err_up = Gamma_high - Gamma_median

# --- Step 1: Propagate uncertainty from rW ---

# Estimate absolute contribution of rW uncertainty
# This is treated as a multiplicative uncertainty impacting the normalization
Gamma_err_rW = Gamma_SM * np.mean(r_samples) * sigma_rW

# Combine rW error with MCMC fit errors (up/down) in quadrature
Gamma_err_total_up = np.sqrt(Gamma_err_up**2 + Gamma_err_rW**2)
Gamma_err_total_down = np.sqrt(Gamma_err_down**2 + Gamma_err_rW**2)

# Format result including rW uncertainty
result_string_total = (
    f"Γ(Z → νν̄) = {Gamma_median:.1f} +{Gamma_err_total_up:.1f} / -{Gamma_err_total_down:.1f} MeV "
    f"(MCMC + rW)"
)
print(result_string_total)

# --- Step 2: Add global systematic uncertainty (e.g. θ = 3.2%) ---

theta_global = 0.032  # 3.2% global multiplicative uncertainty

# Compute absolute contribution of global systematic
Gamma_err_global = Gamma_median * theta_global

# Combine all errors in quadrature (MCMC + rW + θ)
Gamma_err_total_up_all = np.sqrt(Gamma_err_up**2 + Gamma_err_rW**2 + Gamma_err_global**2)
Gamma_err_total_down_all = np.sqrt(Gamma_err_down**2 + Gamma_err_rW**2 + Gamma_err_global**2)

# Final formatted result string
result_string_total_all = (
    f"Γ(Z → νν̄) = {Gamma_median:.1f} +{Gamma_err_total_up_all:.1f} / -{Gamma_err_total_down_all:.1f} MeV "
    f"(MCMC + rW + global systematic)"
)
print(result_string_total_all)
```
